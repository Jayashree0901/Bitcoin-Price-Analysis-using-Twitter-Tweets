```python
import pandas as pd
import numpy as np
import re
from textblob import TextBlob
```

```python
df=pd.read_csv('Bitcoin_tweets.csv')
```

```python
#df=df[df['lang']=='en']
```

```python
def cleanTwt(twt):
  twt=re.sub('#bitcoin', 'bitcoin',twt)
  twt=re.sub('#Bitcoin', 'Bitcoin',twt)
  twt=re.sub('#[A-Za-z0-9]+', '',twt)
  twt=re.sub('\\n', '',twt)
  twt=re.sub('https?:\/\/\S+','',twt)
  twt=re.sub('http?:\/\/\S+','',twt)
  return twt
```

```python
df['text']=df['text'].apply(cleanTwt)
```

```python
df.head(5)
```

```python
def getSubjective(twt):
  return TextBlob(twt).sentiment.subjectivity

def getPolarity(twt):
  return TextBlob(twt).sentiment.polarity
```

```python
df['subjectivity']=df['text'].apply(getSubjective)
df['polarity']=df['text'].apply(getPolarity)
```

```python
df
```

```python
def getSentiment(score):
  if score < 0:
    return 'Negative'
  elif score == 0:
    return 'Neutral'
  else:
    return 'Positive'
```

```python
df['sentiment'] = df['polarity'].apply(getSentiment)
df.head(5)
```

```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
plt.figure(figsize=(8,6))



plt.scatter(df['polarity'][:], df['subjectivity'][:], color='Purple')

plt.title('Sentiment Analysis')
plt.xlabel('Polarity')
plt.ylabel('Subjectivity')
plt.show
```
